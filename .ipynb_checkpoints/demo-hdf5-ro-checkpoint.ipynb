```python
#-----------------------------
# Visualize the mock shears for zs = 1.5
#

make_lensing_mocks.shear_vis_mocks(mocks['xr1'], mocks['xr2'], mocks['sr1'], mocks['sr2'], rmaps['kf0'])
```
